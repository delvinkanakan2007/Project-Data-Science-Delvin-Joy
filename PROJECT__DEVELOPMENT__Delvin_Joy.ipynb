**QUESTION 1** -
**Log File Analyzer**

Develop a Log File Analyzer for simulated system logs.
The program should parse text-based log entries, categorize events, count error types,analyze user activity, and filter logs based on time or severity.
String splitting, pattern matching, and dictionary-based aggregation must be used. The
application should present analytical summaries via a menu-driven interface.

```python


logs = []

def add_log():
    print("Enter log in this format: DATE TIME SEVERITY USER MESSAGE")
    print("Example: 2024-06-19 10:23:45 ERROR user1 Disk failure")
    entry = input("Enter Log Entry: ")

    parts = entry.split(" ", 4)
    if len(parts) < 5:
        print("Invalid log format. Try again.\n")
        return

    date, time, severity, user, message = parts

    log = {
        "date": date,
        "time": time,
        "severity": severity.upper(),
        "user": user,
        "message": message
    }

    logs.append(log)
    print("Log entry added successfully.\n")


def categorize_events():
    categories = {}
    for log in logs:
        severity = log["severity"]
        categories[severity] = categories.get(severity, 0) + 1

    print("\n--- Event Categories ---")
    for category, count in categories.items():
        print(f"{category}: {count}")
    print()


def count_error_types():
    error_messages = {}
    for log in logs:
        if log["severity"] == "ERROR":
            msg = log["message"]
            error_messages[msg] = error_messages.get(msg, 0) + 1

    print("\n--- Error Type Counts ---")
    if not error_messages:
        print("No errors found.")
    for msg, count in error_messages.items():
        print(f"{msg}: {count}")
    print()


def analyze_user_activity():
    user_activity = {}
    for log in logs:
        user = log["user"]
        user_activity[user] = user_activity.get(user, 0) + 1

    print("\n--- User Activity ---")
    for user, count in user_activity.items():
        print(f"{user}: {count} log(s)")
    print()


def filter_by_severity():
    severity = input("Enter severity to filter (INFO/WARNING/ERROR): ").upper()
    print(f"\n--- Logs with severity {severity} ---")
    found = False
    for log in logs:
        if log["severity"] == severity:
            print(f"{log['date']} {log['time']} {log['severity']} {log['user']} {log['message']}")
            found = True
    if not found:
        print("No matching logs found.")
    print()


def filter_by_time():
    date = input("Enter date to filter (YYYY-MM-DD): ")
    print(f"\n--- Logs on {date} ---")
    found = False
    for log in logs:
        if log["date"] == date:
            print(f"{log['date']} {log['time']} {log['severity']} {log['user']} {log['message']}")
            found = True
    if not found:
        print("No matching logs found.")
    print()


# ---------------- Menu Driven Program ----------------
while True:
    print("----- Log File Analyzer -----")
    print("1. Add Log Entry")
    print("2. Categorize Events")
    print("3. Count Error Types")
    print("4. Analyze User Activity")
    print("5. Filter Logs by Severity")
    print("6. Filter Logs by Time")
    print("7. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_log()
    elif choice == "2":
        categorize_events()
    elif choice == "3":
        count_error_types()
    elif choice == "4":
        analyze_user_activity()
    elif choice == "5":
        filter_by_severity()
    elif choice == "6":
        filter_by_time()
    elif choice == "7":
        print("Exiting program. Thank you!")
        break
    else:
        print("Invalid choice. Try again.\n")
```

**QUESTION 2** -
**Course Registration System**

Design a University Course Registration System.
The system should allow creation of courses with credits and seat limits, student
registration, and course enrollment.
Constraints include maximum credit limits per student, prevention of duplicate
registrations, and seat availability enforcement. Summary reports for students and courses
must be generated.

```python
# University Course Registration System (Input Based)

courses = {}     # course_code -> {"name", "credits", "seats", "enrolled"}
students = {}     # student_id -> {"name", "credits_taken", "courses"}

MAX_CREDIT_LIMIT = 24


def add_course():
    code = input("Enter Course Code: ")
    if code in courses:
        print("Course already exists.")
        return
    name = input("Enter Course Name: ")
    credits = int(input("Enter Credits: "))
    seats = int(input("Enter Total Seats: "))
    courses[code] = {"name": name, "credits": credits, "seats": seats, "enrolled": []}
    print(f"Course '{name}' added successfully.\n")


def add_student():
    student_id = input("Enter Student ID: ")
    if student_id in students:
        print("Student already exists.")
        return
    name = input("Enter Student Name: ")
    students[student_id] = {"name": name, "credits_taken": 0, "courses": []}
    print(f"Student '{name}' registered successfully.\n")


def enroll_student():
    student_id = input("Enter Student ID: ")
    if student_id not in students:
        print("Student not found.")
        return

    code = input("Enter Course Code: ")
    if code not in courses:
        print("Course not found.")
        return

    student = students[student_id]
    course = courses[code]

    if code in student["courses"]:
        print("Student already enrolled in this course.")
        return

    if len(course["enrolled"]) >= course["seats"]:
        print("No seats available in this course.")
        return

    if student["credits_taken"] + course["credits"] > MAX_CREDIT_LIMIT:
        print("Credit limit exceeded. Cannot enroll.")
        return

    student["courses"].append(code)
    student["credits_taken"] += course["credits"]
    course["enrolled"].append(student_id)
    print(f"{student['name']} enrolled in {course['name']} successfully.\n")


def student_report():
    student_id = input("Enter Student ID: ")
    if student_id not in students:
        print("Student not found.")
        return
    s = students[student_id]
    print(f"\nStudent Report - {s['name']} (ID: {student_id})")
    print(f"Total Credits: {s['credits_taken']}")
    print("Enrolled Courses:")
    for code in s["courses"]:
        c = courses[code]
        print(f"  {code} - {c['name']} ({c['credits']} credits)")
    print()


def course_report():
    code = input("Enter Course Code: ")
    if code not in courses:
        print("Course not found.")
        return
    c = courses[code]
    print(f"\nCourse Report - {c['name']} (Code: {code})")
    print(f"Credits: {c['credits']}")
    print(f"Seats: {len(c['enrolled'])}/{c['seats']} filled")
    print("Enrolled Students:")
    for sid in c["enrolled"]:
        print(f"  {sid} - {students[sid]['name']}")
    print()


# ---------------- Menu Driven Program ----------------
while True:
    print("----- University Course Registration System -----")
    print("1. Add Course")
    print("2. Add Student")
    print("3. Enroll Student")
    print("4. Student Report")
    print("5. Course Report")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_course()
    elif choice == "2":
        add_student()
    elif choice == "3":
        enroll_student()
    elif choice == "4":
        student_report()
    elif choice == "5":
        course_report()
    elif choice == "6":
        print("Exiting program. Thank you!")
        break
    else:
        print("Invalid choice. Try again.\n")
```
